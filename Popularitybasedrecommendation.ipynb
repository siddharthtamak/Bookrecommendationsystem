# Popularity Based Recommender System

```python
ratings.merge(books, on='ISBN')
```

```python
ratings_with_name= ratings.merge(books, on="ISBN")
```

```python
ratings_with_name.groupby("Book-Title").count()
```

```python
ratings_with_name.groupby("Book-Title").count() ['Book-Rating']
```

```python
ratings_with_name.groupby("Book-Title").count() ['Book-Rating'].reset_index()
```

```python
num_rating_df=ratings_with_name.groupby("Book-Title").count() ['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={"Book-Rating":"num_ratings"},inplace=True)
num_rating_df
```

```python
avg_ratings_pre= ratings_with_name.groupby("Book-Title")
```

```python
ratings_with_name.dtypes
```

```python
ratings_with_name2= ratings_with_name.groupby("Book-Title")
```

```python
ratings_with_name2
```

```python
avg_ratings= ratings_with_name2["Book-Rating"].mean()
avg_ratings
```

```python
avg_ratings= avg_ratings.reset_index()
```

```python
avg_ratings
```

```python
avg_ratings= avg_ratings.rename(columns={"Book-Rating":"avg_ratings"})
```

```python
avg_ratings
```

```python
num_rating_df.merge(avg_ratings, on="Book-Title")
```

```python
popularity_df= num_rating_df.merge(avg_ratings, on="Book-Title")
popularity_df
```

```python
popular_df=popularity_df[popularity_df["num_ratings"]>=220]
```

```python
popular_df
```

```python
popular_df.sort_values("avg_ratings", inplace= True, ascending= False)
```

```python
popular_df
```

```python
popular_df= popular_df.head(50)
```

```python
popular_df
```

```python
popular_df= popular_df.merge(books,on= "Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Year-Of-Publication","Publisher","num_ratings","avg_ratings","Image-URL-M"]]
```

```python
popular_df= popular_df.reset_index()
```
